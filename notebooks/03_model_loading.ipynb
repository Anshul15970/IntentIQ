```python
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch

model_name = "Qwen/Qwen2.5-0.5B-Instruct"

tokenizer = AutoTokenizer.from_pretrained(model_name)

model = AutoModelForCausalLM.from_pretrained(
    model_name,
    torch_dtype="auto",
    device_map="auto"
)

print("Model loaded successfully!")
```

```python
text = "I lost my debit card yesterday."

tokens = tokenizer(text)

print(tokens)
```

```python
token_ids = tokens["input_ids"]

print(token_ids)

print(tokenizer.decode(token_ids))
```

```python
messages = [
    {
        "role": "system",
        "content": """You are an intent classification model.

Your task is to classify the user's query into exactly one of these intents:

- activate_my_card
- age_limit
- apple_pay_or_google_pay
- atm_support
- automatic_top_up
- balance_not_updated_after_cheque_or_bank_transfer
- beneficiary_not_allowed
- ... (add all 77 Banking77 intents)

Rules:
1. Return ONLY the intent label.
2. Do not explain.
3. Do not write a sentence.
4. Output exactly one label.
"""
    },
    {
        "role": "user",
        "content": "I lost my debit card yesterday."
    }
]

text = tokenizer.apply_chat_template(
    messages,
    tokenize=False,
    add_generation_prompt=True
)

inputs = tokenizer(text, return_tensors="pt").to(model.device)

outputs = model.generate(
    **inputs,
    max_new_tokens=200
)

generated_tokens = outputs[0][inputs["input_ids"].shape[1]:]

response = tokenizer.decode(
    generated_tokens,
    skip_special_tokens=True
)

print(response)
```
